```python
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
# Create a model instance and call it directly
model = OpenAI(model_name="gpt-4o", temperature=0)
print(model("The sky is"))
```

```python
# Environment & model checks
import os
print("OPENAI_API_KEY set:", bool(os.environ.get("OPENAI_API_KEY")))

# Optional: alternative for chat models
try:
    from langchain.chat_models import ChatOpenAI
    from langchain.schema import HumanMessage
    chat = ChatOpenAI(model_name="gpt-4o", temperature=0)
    print(chat([HumanMessage(content="The sky is")])[0].content)
except Exception:
    # ChatOpenAI not available or not configured â€” ignore
    pass
```
